*  **EEG:**Teniendo en cuenta que los electroencefalogramas nos ayudan a registrar la actividad eléctrica del cerebro, procederemos a conceptualizar cada una de las bandas espectrales presentadas en las EEG, para lograr una  mayor claridad frente a la escogencia de las caracteristicas mas relevantes de dicha señal. 
      * **Banda Delta:** Definidas hasta los 3,5 Hz.  Son ondas de baja frecuencia y alta intensidad (centenares de μV). Tienen lugar en niños de corta edad y en adultos únicamente en estado de sueño profundo, inconsciencia o situaciones que aumente la presión intracraneal como tumores
cerebrales.
      * **Banda Theta:**De los 3,5 a 7,5 Hz. Estas ondas de amplitud inferior a 20 μV se dan durante el proceso de maduración en toda la corteza cerebral, aunque predominan en la región occipital y temporal y es más rápida en la zona frontal. Es clasificada como "actividad lenta", se presenta con normalidad en niños de hasta 14 años y anormalidad en adultos despiertos. Puede verse como una manifestación de lesiones o trastornos difusos. 
      * **Banda Alfa:** Definida entre los 7.5 y 12.5 Hz. Término usado para ráfagas de 20-100 μV de amplitud y gran
periodicidad. Estas ondas se asocian a periodos de relajación, de inactividad, y son muy patentes en ausencia de estímulos visuales (cuando se cierran los ojos).Se desaparece al abrir los ojos o alertar por cualquier mecanismo (pensar, calcular).
Existe mucha variabilidad interpersonal en el ritmo alfa. Hay personas sin ritmo alfa y otras con ritmo alfa permanente, aunque suele observarse regularmente en adultos relajados y es una de las ondas  presentes la mayor parte de la vida (después de los 13 años)
      * **Banda Beta:**de 12,5 a 30 Hz. Son señales de muy baja amplitud, por debajo de 20 μV, son bastante comunes y predominan durante la edad adulta. Suelen subdividirse en beta baja (suele localizar en los lóbulos frontal y occipital), media y alta (menos localizados).
Más irregular que el ritmo alfa, se asocia a actividad psicofísica, estados de agitación, alerta, ansiedad o la actividad mental que se realiza en la resolución de problemas.
      * **Banda Gamma:** Se referiere al rango de frecuencias de 30 a 50 Hz 
Esta banda tiene un pico de resonancia, similar a los ritmos alfa, cercano a los 40Hz al que se le da el nombre de épsilon, que se asocia a la actividad mental abstracta que interviene en percibir como único, por ejemplo, el olor, los rasgos de la cara, la personalidad y la voz de una persona.

  Con las anteriores definiciones, tenemos una base para caracterizar nuestra actividad cerebral, para la cual se tendrá en cuenta:

```

def media(data):
  media= np.mean(data)
  #print('Media =',media)
  return media

def varianza(data):
  varianza= np.var(data)
  #print('Varianza =',varianza)
  return varianza

def energia(señal):
  energia= np.sum(np.power(señal,2))
  #print('Energía =',energia)
  return energia

def media_deriv(data):
  derivada = np.diff(data)
  media= np.mean(derivada)
  #plt.plot(derivada)
  #print('Derivada promedio =',media)
  return media

def desvi_estand(data):
  desvi= np.std([data], ddof=1)
  #print('Desviación Estandar =',desvi)
  return desvi

def media_deri_neg(data):
  n=[]
  derivada = np.diff(data)
  for i in derivada:
    if i < 0:
      n=[i]
      median= np.mean(n)
  return median

def psd(data):
  fs = 128 # fs
  f, Pxx_fil = signal.welch(data, fs, nperseg=64)
  #plt.semilogy(f, Pxx_fil)
  #plt.ylim([0.5e-3, 1])
  #plt.xlabel('frequency [Hz]')
  #plt.ylabel('PSD [V**2/Hz]')
  #plt.show()
  return (Pxx_fil)  
  
def potencia(data):
  pot = np.sum(data)
  #print("P",pot)
  return pot

def lowpass_filter(data, fcorte, fs, orden):
    nyq = 0.5*fs
    fcorte_nor = fcorte / nyq
    # Get the filter coefficients 
    b, a = signal.butter(orden, fcorte_nor, btype='low', analog=False)
    y = signal.filtfilt(b, a, data)
    return y

def highpass_filter(data, fcorte, fs, orden):
    nyq = 0.5*fs
    fcorte_nor = fcorte / nyq
    # Get the filter coefficients 
    b, a = signal.butter(orden, fcorte_nor, btype='high', analog=False)
    y = signal.filtfilt(b, a, data)
    return y

def passband_filter(data, fmin, fmax, fs, orden):
    nyq = 0.5*fs
    fmin_nor = fmin / nyq
    fmax_nor = fmax / nyq
    b, a = signal.butter(orden, [fmin_nor, fmax_nor], btype='band', analog=False)
    y = signal.filtfilt(b, a, data)
    return y

def filtro_soslp(data, orden, fc):
  sos = signal.butter(orden, fc, 'lp', fs=128, output='sos')
  filtered = signal.sosfilt(sos, data)
  #plt.plot(time, filtered)
  #plt.title('Banda 0-0.1Hz')
  #plt.xlabel('Time [seconds]')
  #plt.show()
  return filtered

def filtro_sospb(data, orden, fcmin, fcmx):
  sos = signal.butter(6, [fcmin, fcmx], 'bp', fs=128, output='sos')
  filtered2 = signal.sosfilt(sos, data)
  #plt.plot(time, filtered2)
  #plt.title('Banda entre 0.1Hz y 0.2Hz')
  #plt.xlabel('Time [seconds]')
  #plt.show()
  return filtered2  


def tasa(data):
  peaks, _ = find_peaks(data, prominence=np.mean(ojov))
  np.diff(peaks)
  #plt.plot(data)
  #plt.plot(peaks, data[peaks], "x")
  #plt.show()
  return (len(peaks))
```

```
labels = db['labels']

Valence = labels[:,0]
Arousal = labels[:,1]


negA = np.logical_and(Arousal>=1,Arousal<=3)
negV = np.logical_and(Valence>=1,Valence<=3)
neutralA = np.logical_and(Arousal>3,Arousal<7)
neutralV = np.logical_and(Valence>3,Valence<7)
posA = Arousal>=7
posV = Valence>=7
Etiquetas = np.zeros((40,1),dtype=int)
# Emocion Negativa negA y negV
Etiquetas[np.where(np.logical_and(negA,negV))] = 1
# Emocion Neutral neutralA y neutralV
Etiquetas[np.where(np.logical_and(neutralA,neutralA))] = 2
# Emocion Positiva posA y posV
Etiquetas[np.where(np.logical_and(posA,posV))] = 3
# Emocion Stress posA y negV
Etiquetas[np.where(np.logical_and(posA,negV))] = 4
# Emocion Calm negA y posV
Etiquetas[np.where(np.logical_and(negA,posV))] = 5
# Emocion Tristeza neutralA y negV
Etiquetas[np.where(np.logical_and(neutralA,negV))] = 6
# Emocion Satisfecho (Pleasant) neutralA y posV
Etiquetas[np.where(np.logical_and(neutralA,posV))] = 7
# Emocion Tranquilidad negA y neutralV
Etiquetas[np.where(np.logical_and(negA,neutralV))] = 8
# Emocion Sorpresa posA y neutralV
Etiquetas[np.where(np.logical_and(posA,neutralV))] = 9



Eti = Etiquetas.reshape(40)

print(Eti)


Etiquetas_name = []

for i in range (0, len(Eti)):

  if (Eti[i] == 1):
    Etiquetas_name.append('Negativa')

  elif (Eti[i] == 2):
    Etiquetas_name.append("Neutral")

  elif (Eti[i] == 3):
    Etiquetas_name.append("Positivo")

  elif (Eti[i] == 4):
    Etiquetas_name.append("Estres")

  elif (Eti[i] == 5):
    Etiquetas_name.append("Calma")

  elif (Eti[i] == 6):
    Etiquetas_name.append("Tristeza")

  elif (Eti[i] == 7):
    Etiquetas_name.append("Satisfech")

  elif (Eti[i] == 8):
    Etiquetas_name.append("Tranquilidad")

  elif (Eti[i] == 9):
    Etiquetas_name.append("Sorpresa")

print(Etiquetas_name)
```
